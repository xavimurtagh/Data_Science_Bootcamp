```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# opening insurance and viewing first 5 rows
insurance = pd.read_csv("Insurance.csv")
insurance.head()
```

```python
insurance['age'] = insurance['age'].astype('int')
insurance['charges'] = insurance['charges'].astype('float')
```

```python
# setting x and y
x = insurance.iloc[:,:1].values
y = insurance.iloc[:,-1].values
```

```python
# getting linear regression model
model = LinearRegression()
model.fit(x,y)
```

```python
# plot scatter graph
plt.scatter(x,y,color = 'b')
plt.show()
```

From what i can see in this graph i predict that as age increases so does charges

```python
# drawing line of best to check if prediction matches trend
y_pred = model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.ylabel('Charges')
plt.xlabel('Age')
plt.show()
```

```python
# making prediction
unk_x = [[82]]
print("82 year old should pay:", model.predict(unk_x))
```

As i thought, as a persons age increases so does their charges
